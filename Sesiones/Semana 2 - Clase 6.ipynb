# Gr√°ficas de contorno

```python
import numpy
import matplotlib.pyplot as pyplot
```

```python
x = [1, 2, 3]
y = [4, 5, 6]

X, Y = numpy.meshgrid(x, y)

X, Y
```

$$
f(x, y) = cos(x^2 + y^2)
$$

```python
Z = numpy.cos(X ** 2 + Y ** 2)

Z
```

```python
pyplot.imshow(Z)
```

```python
Xp, Yp = numpy.meshgrid(numpy.linspace(0, 10), numpy.linspace(0, 10))

Z = numpy.cos(Xp ** 2 + Yp ** 2)

pyplot.imshow(Z, cmap="magma", extent=[0, 10, 0, 10])
pyplot.colorbar()
pyplot.title("f(x, y) = cos(x^2 + y^2)")
pyplot.show()
```

```python
Xp, Yp = numpy.meshgrid(numpy.linspace(0, 10, 200), numpy.linspace(0, 10, 200))

Z = numpy.cos(Xp ** 2 + Yp ** 2)

pyplot.imshow(Z, cmap="magma", extent=[0, 10, 0, 10])
pyplot.colorbar()
pyplot.title("f(x, y) = cos(x^2 + y^2)")
pyplot.show()
```

$$
f(x, y) = \begin{cases}
\frac{sin(x^2 + y^2)}{x^2 + y^2} \qquad x^2 + y^2 > 1 \\
\frac{1}{2} \qquad x^2 + y^2 <= 1
\end{cases}
$$

```python
Xp, Yp = numpy.meshgrid(numpy.linspace(-4, 4, 200), numpy.linspace(-4, 4, 200))

Rp = Xp ** 2 + Yp ** 2

Z = (Rp > 1) * numpy.sin(Rp) / Rp + (Rp <= 1) * 0.5

# Z = Rp * 0
# Z[Rp > 1] = numpy.sin(Rp) / Rp
# Z[Rp <= 1] = 1 / 2

pyplot.imshow(Z, cmap="magma", extent=[-4, 4, -4, 4])
pyplot.colorbar()
pyplot.title("Contorno")
pyplot.show()
```
