## Ozono ($O_3$) por zona

```python
indices_O3 = indices[[
    "Noroeste Ozono",
    "Noreste Ozono",
    "Centro Ozono",
    "Suroeste Ozono",
    "Sureste Ozono",
]].copy()

indices_O3.columns = indices_O3.columns.map(lambda s: s.split(" ")[0])

indices_O3.head()
```

```python
z1 = indices_O3["Noroeste"].copy()
```

```python
indices_O3.isna().sum()
```

```python
for column in indices_O3.columns:
    indices_O3[column] = indices_O3[column].fillna(indices_O3[column].median())

indices_O3.isna().sum()
```

```python
indices_O3.describe()
```

```python
z2 = indices_O3["Noroeste"].copy()

z1.plot.density()
z2.plot.density()
```

```python
indices_O3.index = pandas.to_datetime(indices.index + pandas.to_timedelta(indices["Hora"] - 1, unit="h"))

indices_O3.head()
```

```python
indices_O3_periodo = indices_O3.loc["2017-01-01":"2017-01-03"]

indices_O3_periodo.head()
```

```python
for column in indices_O3_periodo.columns:
    seaborn.lineplot(indices_O3_periodo, x=indices_O3_periodo.index, y=column, 
                     label=column, linestyle="solid")

pyplot.title("Concentraciones de $O_3$ por zona")
pyplot.xlabel("Fecha y hora")
pyplot.ylabel("Concentración de $O_3$ $(\\mu g/m^3)$")
pyplot.xticks(rotation=90)

pyplot.show()
```

## Remuestreo temporal (por día)

```python
indices_O3_dia = indices_O3.resample("D").mean()

indices_O3_dia.head()
```

```python
for column in indices_O3_dia.columns:
    seaborn.boxplot(indices_O3_dia)

pyplot.title("Concentraciones de $O_3$ por zona")
pyplot.xlabel("Fecha y hora")
pyplot.ylabel("Concentración de $O_3$ $(\\mu g/m^3)$")
pyplot.xticks(rotation=90)

pyplot.show()
```

```python
indices_O3_dia.mean(axis=1).head()
```

```python
indices_O3_dia.mean(axis=1).plot(marker=".", linestyle="dotted")
```

# Discretización

```python
corredores = pandas.read_csv("Corredores.csv")

corredores.head()
```

```python
corredores.info()
```

```python
corredores.describe()
```

```python
bins = [0, 17, 29, 39, 49, 59, 90]

grupos = pandas.cut(corredores["Edad"], bins) # right=True

grupos.head()
```

```python
grupos.value_counts()
```

```python
grupos = pandas.cut(corredores["Edad"], bins, labels=[
    "<18", 
    "18 a 29 años", 
    "30 a 39 años", 
    "40 a 49 años", 
    "50 a 59 años", 
    "60 o más", 
])

corredores["Grupo_edad"] = grupos

corredores.head()
```

```python
seaborn.barplot(corredores, x="Tiempo", y="Grupo_edad", hue="Genero")
```

```python
seaborn.boxplot(corredores, x="Grupo_edad", y="Tiempo", hue="Genero")
```

## Gráficas de contrastes

```python
figure, axis = pyplot.subplots(nrows=2, sharex=True, figsize=(10, 6), height_ratios=[1, 3])

seaborn.boxplot(corredores, x="Edad", ax=axis[0], color="skyblue")
axis[0].set_title("Concentración de las edades")

seaborn.histplot(corredores, x="Edad", bins=30, kde=True, ax=axis[1], color="skyblue")

pyplot.tight_layout()
```
