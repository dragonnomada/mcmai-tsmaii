## Calcular los múltiplos de 3 y 5 del 1 al 10,000

```python
multiplos = []

for n in range(1, 10001):
    if n % 3 == 0 and n % 5 == 0:
        multiplos.append(n)
```

Primeros 10 múltiples de 3 y 5

```python
multiplos[:10]
```

Últimos 10 múltiplos de 3 y 50

```python
multiplos[-10:]
```

Total de múltiplos de 3 y 5

```python
len(multiplos)
```

Suma de los múltiplos de 3 y 5

```python
sum(multiplos)
```

Frecuencia de los múltiplos en cada 100 unidades

```python
import matplotlib.pyplot as pyplot

x = range(1, 10001)

def esMultiplo(xi):
    if xi in multiplos:
        return 1
    else:
        return 0

y = [esMultiplo(xi) for xi in x]

pyplot.plot(x, y, "r.")
pyplot.show()
```

Escribe un programa que evalúe la siguiente función

$$
f : \mathbb{Z} \rightarrow \mathbb{R}
$$

Definida por:

$$
f(x) = \begin{cases}
x^3 \qquad x < 0, \\
e^x \qquad 0 \le x \le 10, \\
\sqrt{x + 3} \qquad x > 10, \\
\end{cases}
$$

Para $x \in [-15, 15]$

```python
from math import exp, sqrt

X = []
Y = []

print("   x    |      y     ")
print("-" * 20)
for x in range(-15, 16):
    if x < 0:
        y = x ** 3
    elif x >= 0 and x <= 10:
        y = exp(x)
    elif x > 10:
        y = sqrt(x + 3)
    print(f"{x:^7d} | {y:^10.2f}")
    X.append(x)
    Y.append(y)
```

```python
pyplot.plot(X, Y, color="royalblue", marker=".", linestyle="--")
pyplot.show()
```

# Ciclo `while`

```python
x = 50
total = 0

while x > 0:
    if total > 700:
        break
    print(total)
    total += x

print("-" * 10)
print(total)
```

### Ejercicio: Solicitar la edad y calcular el grupo

```python
while True:
    edad = int(input("Dame la edad:"))
    if edad == 0:
        break
    elif edad < 18:
        print(f"Edad: {edad} -> Menor de edad")
    elif edad >= 18 and edad <= 29:
        print(f"Edad: {edad} -> Jóven adulto")
    elif edad >= 30 and edad <= 44:
        print(f"Edad: {edad} -> Adulto jóven")
    elif edad >= 45 and edad <= 59:
        print(f"Edad: {edad} -> Adulto maduro")
    else:
        print(f"Edad: {edad} -> Adulto mayor")
```
