```python
import numpy
import pandas

import matplotlib.pyplot as pyplot
import seaborn
```

```python
nacimientos = pandas.read_excel("nacimientos.xlsx")

nacimientos.head()
```

## Agrupamiento y agregaci√≥n

```python
reporteRegiones = nacimientos.groupby(by="Regiones")[["Mujeres", "Hombres", "No_esp", "Total"]].sum()

reporteRegiones
```

```python
reporteRegiones[["Total"]].plot.barh()
```

```python
colores = ["cornflowerblue", "tomato", "green", "crimson", "grey"]

pyplot.barh(reporteRegiones.index, reporteRegiones["Total"], color=colores)
pyplot.title("Total de nacimientos por region")
pyplot.show()
```

## Fragmentos de datos (chunksize)

```python
carpetas = pandas.read_csv("carpetasFGJ.csv", chunksize=1_000)

total = pandas.Series([], dtype="int64")

for chunk in carpetas:
    total = total.add(chunk["alcaldia_hecho"].value_counts(), fill_value=0)
    total = total.sort_values(ascending=False)

delitos = pandas.DataFrame(total, columns=["Frecuencia"])

delitos
```

### Colonias de mayor frecuencia

```python
carpetas = pandas.read_csv("carpetasFGJ.csv", chunksize=1_000)

total = pandas.Series([], dtype="int64")

for chunk in carpetas:
    total = total.add(chunk["colonia_hecho"].value_counts(), fill_value=0)
    total = total.sort_values(ascending=False)

delitos = pandas.DataFrame(total, columns=["Frecuencia"])

delitos
```

```python
delitos.head(15).sort_values(by="Frecuencia").plot.barh()
```

```python
delitos.to_csv("delitos_colonia.csv", sep="|", index=False)
```
